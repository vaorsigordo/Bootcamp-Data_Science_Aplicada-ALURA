### Funções de tabelas de frequência e percentual e crosstab

```
def tabela_frequencia_percentual(dados_coluna,  dicionario_respostas, titulo_indice):
  """
  Retorna um DataFrame pandas expondo a frequência total e o percentual (normalizado pelo total da coluna) na coluna selecionada.

  Parâmetros
  ----------
  coluna,  dicionario_respostas, titulo_indice
  dados["  "], {  : "  ",  : "  ",  : "  ",  : "  "}, "  "
  ----------
  """
  frequencia_regiao = dados_coluna.value_counts()
  percentual = dados_coluna.value_counts(normalize = True) * 100
  respostas = dicionario_respostas
  # criando o DataFrame
  dist_freq = pd.DataFrame({"Frequência": frequencia_regiao, "Percentual": percentual.round(2)})
  # renomeando o index com o dicionário
  dist_freq.rename(index = respostas, inplace = True)
  # renomeando o título do índice
  dist_freq.rename_axis(titulo_indice, axis=1, inplace = True)
  
  return(dist_freq)
```

```
# DataFrame crosstab frequência e percentual

def tabela_crosstab_frequencia_percentual(indices, colunas, dicionario_indice0, dicionario_coluna, normalizacao, dicionario_indice1):
  """
  Retorna um DataFrame pandas expondo a frequência total e o percentual (normalizado pelo total da coluna) 
  na coluna selecionada de dados cruzados.

  Parâmetros
  ----------
  (indice, coluna, dicionario_indice, dicionario_coluna, normalizacao)

  o índices pode ser um string ou pode ser uma lista de dois valores:
  
  indice = dados_alunos["  "] ou [dados_alunos["  "], dados_alunos["  "]],
  coluna = dados_alunos["  "],
  dicionario_indice = {  : "  ",  : "  ",  : "  ",  : "  "}, 
  dicionario_coluna = {  : "  ",  : "  ",  : "  ",  : "  "},
  "columns" ou "index"
  ----------
  """

  #DataFrames de frequência
  frequencia_cruzada_frequencia = pd.crosstab(index =  indices, columns= colunas)
  if type(indices) == list:
    frequencia_cruzada_frequencia.rename(index = dicionario_indice0, level = 0, inplace = True)
    frequencia_cruzada_frequencia.rename(index = dicionario_indice1, level = 1, inplace = True)
  else:
    frequencia_cruzada_frequencia.rename(index = dicionario_indice0, level = 0, inplace = True)

  frequencia_cruzada_frequencia.rename(columns = {chave: f'Frequência - {valor}' for (chave, valor) in dicionario_coluna.items()},level = 0, inplace = True)

  #DataFrames de percentual
  frequencia_cruzada_percentual = pd.crosstab(index =  indices, columns= colunas, normalize = normalizacao)*100
  if type(indices) == list:
    frequencia_cruzada_percentual.rename(index = dicionario_indice0, level = 0, inplace = True)
    frequencia_cruzada_percentual.rename(index = dicionario_indice1, level = 1, inplace = True)
  else:
    frequencia_cruzada_percentual.rename(index = dicionario_indice0, level = 0, inplace = True)
  frequencia_cruzada_percentual.rename(columns = {chave: f'Percentual (%) - {valor}' for (chave, valor) in dicionario_coluna.items()},level = 0, inplace = True)

  #juntando os dois DataFrames o de frequencia e o de percentual
  frequencia_cruzada = pd.concat([frequencia_cruzada_frequencia, frequencia_cruzada_percentual.round(2)], axis=1)
  
  return(frequencia_cruzada)
```

### Funções de plotagem

```
def plotagem_hist1(x, dados, ymax, ymin, titulo, labelx, xlimite_min, xlimite_max):
  """
  Retorna um gráfico de histograma com medidas de centralidade e dispersão plotadas

  Parâmetros
  ----------
  x, dados, ymax, ymin, titulo, labelx, xlimite_min, xlimite_max
  (" ", dados, 0, 60, "  ", "  ", 0, 680)
  ----------
  """
############################ plotar o gráfico 1 ################################
  ax = sns.histplot(x = x, data = dados, kde = True)
########################## Estética do gráfico 1 ###############################
  plt.title(titulo, fontsize = 16, fontweight = "bold") # adiciona título
  plt.xlabel(labelx,fontsize = 14, fontweight = "bold") #nomeia x
  plt.xlim(xlimite_min,xlimite_max)
  plt.tick_params(axis="x", labelsize=12) #modifica o tamanho do tick em x
  plt.tick_params(axis="y", labelsize=12) #modifica o tamanho do tick em y
  plt.grid(True, color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
  #plotar medida de centralidade e dispersao e adicionar legenda destes cálculos
  media = dados[x].mean()
  std_media = [media + dados[x].std(), media - dados[x].std()]
  ax.vlines(dados[x].mean(), ymin=ymin, ymax=ymax, color='r',label="Média", linestyle="--", linewidth=2.5)
  plt.axvspan(std_media[1], std_media[0], facecolor='0.7', alpha=0.5, label="Desvio Padrão")
  ax.vlines(dados[x].median(), ymin=ymin, ymax=ymax, color='k', label='Mediana', linestyle="--", linewidth=2.5)
  ax.vlines(dados[x].mode(), ymin=ymin, ymax=ymax, color='g', label='Moda', linestyle="--", linewidth=2.5)
  plt.legend()
```

```
def plotagem_boxplot( x, y, dados, xlabel, ylabel, titulo, lista_yticks_string):
  """
  Retorna um gráfico de histograma com medidas de centralidade e dispersão plotadas

  Parâmetros
  ----------
  x, y, dados, xlabel, ylabel, titulo, lista_yticks_string
  ("x", "y", dados,"xlabel", "ylabel", "titulo", ["string", "string", "string" ...])
  ----------
  """
############################# plotar o gráfico #################################
  ax = sns.boxplot(x = x, y = y , data = dados, orient = "h")
########################### Estética do gráfico ################################
  ax.figure.set_size_inches(12, 6)
  ax.set_xlabel(xlabel,fontsize = 14, fontweight = "bold")
  ax.set_ylabel(ylabel,fontsize = 14, fontweight = "bold")
  plt.tick_params(axis="x", labelsize=12) #modifica o tamanho do tick em x
  plt.tick_params(axis="y", labelsize=12) #modifica o tamanho do tick em y
  ax.set_title(titulo, fontsize = 16, fontweight = "bold")
  #trocar os números do eixo y para string
  y = (dados[y].unique())-1
  nomeia_yticks = lista_yticks_string
  plt.yticks(y, nomeia_yticks)
  ax;
```

### Função que calcula os valores de centralidade e dispersão
- Média, Moda, Mediana e Desvio Padrão

```
def calcular_centralidade_e_dispersao_da_distribuição(x, dados, titulo):
  """"
  Esta função calcula os valores de centralidade e dispersão: Média, Moda, Mediana, Desvio Padrão, Primeiro e Terceiro Quartio.
  Parâmetros
  ----------
  x, dados, titulo
  ("x", dados, "titulo")
  ----------

  """
  media = round((dados[x].mean()),2)
  mediana = round((dados[x].median()), 2)
  moda = dados[x].mode().tolist()
  moda_arredondada = list(map(lambda x: round(x, 2), moda))
  desvio_padrao = [media - dados[x].std(), media + dados[x].std()]
  desvio_padrao_arredondado = list(map(lambda x: round(x, 2), desvio_padrao))
  primeiro_quartio = (round(dados[x].quantile(0.25),2))
  terceiro_quartio = (round(dados[x].quantile(0.75),2))
  print(f'\n{titulo}\nMédia = {media}\nModa = {moda_arredondada}\nMediana = {mediana}\nDesvio Padrão = {desvio_padrao_arredondado}\nPrimeiro Quartio = {primeiro_quartio}\nTerceiro Quartio = {terceiro_quartio}')
```

## CÁLCULO DE IMC

IMC = massa (em kg) / altura ao quadrado (em metros)

O IMC é um padrão internacional utilizado para avaliar grandes populações, embora não seja o método mais preciso para avaliar individualmente a gordura e a massa magra.

```
# Cálculo de IMC
# "VB17003" - Peso do estudante, incluindo os imputados, com uma casa decimal (MASSA em Kg)
# "VB17004" - Altura do estudante, incluindo os imputados, com uma casa decimal (MASSA em cm) preciso transformar para metros, por isto divido por 100
dados_alunos["IMC"] = dados_alunos["VB17003"]/((dados_alunos["VB17004"]/100)**2)
dados_alunos["IMC"]


# Criando classificações do IMC
# classificações já conhecidas de IMC
classes = [0, 18.5, 25, 30, 60]
# nome destas classificações
labels = ["abaixo", "normal", "sobrepeso", "obesidade"]

# recortando (cut) em intervalos das classes definidas
dados_alunos["CLASSE_IMC"] = pd.cut(x = dados_alunos["IMC"], bins = classes, labels=labels, include_lowest = True)
 
dados_alunos[["IMC", "CLASSE_IMC"]].head()
```

###**TESTE DE HIPÓTESES**

A seguir será feito um cálculo estatístico para a distribuição do IMC de todos os escolares:

Será testado se a distribuição do IMC é a mesma ou diferente dos escolares que NUNCA passaram fome nos últimos 30 dias para os escolares que SEMPRE passaram fome nos últimos 30 dias.

O teste é feito para saber se há diferenças estatística relevante na distribuição no IMC dos dois grupos.

Hipótese nula $H_0$

Hipótese alternativa $H_1$

***


***
Para decidir que teste de hipóteses será utilizado primeiramente deve-se testar se a curva desta distribuição é proveniente de uma distribuição normal ou não. 

Portanto, (1) primeiramente feito o teste de normalidade  (2) na sequência será plotado as curvas,e em seguida (3) decidir o teste e testar as hipóteses.
***

(1) **TESTE DE NORMALIDADE**

O normaltest testa a hipótese nula $H_0$ de que a amostra é proveniente de uma distribuição normal. 

Rejeitamos $H_0$ se o valor $p \leq 0,05$ (o valor de 0,05 é o valor que atribuimos a nossa significância) e aceitamos a hipótese alternativa $H_1$

```
from scipy.stats import normaltest
#distribuicao_IMC_nunca = dados_alunos.query("VB02024 == 1")["IMC"]
distribuicao_IMC_nunca = dados_nunca_passou_fome["IMC"]
#distribuicao_IMC_sempre = dados_alunos.query("VB02024 == 5")["IMC"]
distribuicao_IMC_sempre = dados_sempre_passou_fome["IMC"]
_, p_value_normaltest_nunca = normaltest(distribuicao_IMC_nunca)
_, p_value_normaltest_sempre = normaltest(distribuicao_IMC_sempre)
print(p_value_normaltest_nunca)
print(p_value_normaltest_sempre)
```

***

Observe que ele retornou um pvalue é um valor menor que 0.05 nos dois casos, assim verificação nula diz que nenhuma das duas distribuições são provenientes de distribuição normal.

(3) **TESTE NÃO PARAMÉTRICO RANKSUMS**

(3) Agora vamos observar se as distribuições apresentam diferenças estatisticamente:

Como a distribuição dos IMCs não são distribuição normal então será utilizado um teste não paramétrico, será utilizado o teste não paramétrico $Ranksums$.

Hipótese nula $H_0$: A distribuição do IMC é a mesma para os dois grupos de escolares.

Hipótese alternativa $H_1$: A distribuição do IMC é diferente para os dois grupos de escolares.

[3] Disponível em: https://www.ime.unicamp.br/~dias/Ch10.wilcoxon.pdf Acesso em: 08 de jan. de 2021.


```
from scipy.stats import ranksums

ranksums(distribuicao_IMC_nunca, distribuicao_IMC_sempre)
```

Antes de analisarmos é importante resultar o seguinte: o p value menor que 0.05 (valor atribuido a significância) significa que eu quero um nível de confiança igual a 95%

Analisando o valor do p value do teste de hipóteses $Ranksums$:

Observe que ele retornou um p value é um valor menor que 0.05, desta forma recuso a minha hipótese zero, significando que estas distribuições são provenientes de distribuições diferentes.

Em outras palavras foi rejeitei a hipótese nula $H_0$ (p value menor que 005), logo a distribuição do IMC dos dois grupos de escolares analisados (SEMPRE e NUNCA passaram fome, nos últimos 30 dias) são estatisticamente diferentes -  pvalue menos que 0.05 para o ranksums test.
