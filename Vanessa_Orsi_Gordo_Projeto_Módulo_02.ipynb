## Resumo:

A cobertura vacinal no Brasil vem atingindo índices preocupantes ao longo do últimos anos, com este estudo foi possível concluir que a cobertura vacinal caiu a níveis que se aproximam da cobertura vacinal de 1996, em termos científicos esta cobertura é um retrocesso nos avanças alcançados pelo Programa Nacional de Imunização, deixando a população mais suscetível a contrair doenças que antes eram erradicadas. Foi possível identificar os anos em que houve um maior desabastecimento e o problema da não homogeneidade da cobertura vacinal.

Neste projeto será analisado a cobertura vacinal, será tentado identificar qual estados apresenta uma maior cobertura vacinal e se algum estado apresenta maior ou menor cobertura. Também será analisado se apresentam mais casos de sarampo e febre amarela, além disso, pretende-se identificar algumas possíveis causas para a não vacinação.

```
dados_sage_sarampo_copy = dados_sage_sarampo.copy()
```

```
# Adicionar a coluna de região no DataFrame
sudeste = ["SP","MG","RJ","ES"]
norte = ["PA","AM","RO","TO","AC","AP","RR"]
nordeste = ["AL","BA","CE","MA","PB","PE","PI","RN","SE"]
centro_oeste = ["DF","GO","MT","MS"]
sul = ["PR","SC","RS"]

lista_uf = dados_sage_sarampo_copy['uf'].tolist()

def region(lista_uf):
    """
    Esta função retorna o nome da regiao de um estado
    """
    
    if lista_uf in sudeste:
        return 'Sudeste'
    elif lista_uf in norte:
        return 'Norte'
    elif lista_uf in nordeste:
        return 'Nordeste'
    elif lista_uf in centro_oeste:
        return 'Centro-Oeste'
    elif lista_uf in sul:
        return 'Sul'

    else:
        return None
dados_sage_sarampo_copy['regiao'] = [region(i) for i in lista_uf]
dados_sage_sarampo_copy
```

# **CONCLUSÃO**

Há uma enorme necessidade de intensificar as ações em locais com baixas coberturas, e locais onde há surtos da doença, como a febre amarela e o sarampo.

Neste estudo podemos verificar que a cobertura vacinal caiu a preocupantes níveis de cobertura, uns dos anos mais críticos analisados é o de 2019, cuja a cobertura está extremamente abaixo da meta. Para se ter dimensão do problema a cobertura atual de vacinação se aproxima da cobertura vacinal apresentada em 1996, em termos científicos esta cobertura é um retrocesso nos avanças alcançados pelo Programa Nacional de Imunização, deixando a população mais suscetível a contrair doenças antes erradicadas.

Em 2015 - 2016 foi possível verificar uma enorme queda na cobertura vacinal e foi encontrado em reportagens, que Ministério da Saúde admitiu o desabastecimento [9]. Um outro resultado obtido é o da falta das vacinas a serem aplicadas, é um problema real e atingiu, no segundo semestre de 2018, em torno de 5,92% dos não vacinados no Ceará e 7,81% no Paraná.

Também foi observado que embora um estado tenha uma cobertura vacinal considerável, ou próxima da meta, a não homogeneidade desta cobertura vacinal nos municípios, podem deixar regiões ou locais mais suscetíveis a doenças como o caso encontrado de Febre Amarela em Minas Gerais nestes últimos anos. No caso deste estudo podemos ver um alto do número de infectados em 2020.

Quanto ao movimento anti vacinação, há estudos que mostram que as "fake news" possuem maior engajamento (13,5% dos links) se espalhando mais facilmente nas redes sociais [18]. Como já observado este movimento é antigo, um exemplo é o de 1885 durante a epidemia de varíola em Montreal onde haviam panfletos de movimento anti vacinação [19]. Desde então, vários argumentos são usados para a não vacinação.
